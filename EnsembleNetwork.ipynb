```python
import tensorflow as tf
from tensorflow.keras import layers, models, regularizers
import numpy as np
from sklearn.metrics import accuracy_score
import sys
from tensorflow.keras import models, layers
import numpy as np
from sklearn.metrics import accuracy_score


from tensorflow.keras.models import load_model


load_path = '/content/gdrive/My Drive/cnn_model3.keras'


reloaded_model1 = load_model(load_path)



load_path = '/content/gdrive/My Drive/cnn_sgd1.keras'


reloaded_model2 = load_model(load_path)




load_path = '/content/gdrive/My Drive/cnn_sgd4.keras'

reloaded_model3 = load_model(load_path)

print("1. Loading and Preprocessing of dataset CIFAR-10...")
(_, _), (X_test, y_test) = tf.keras.datasets.cifar10.load_data()


X_test = X_test.astype('float32') / 255.0


y_test_one_hot = tf.keras.utils.to_categorical(y_test, num_classes=10)
y_test_flat = y_test.flatten()



models_to_ensemble = []
print("\n2. Model Loading (.keras)...")

reloaded_model1.trainable = False
reloaded_model2.trainable = False
reloaded_model3.trainable = False
models_to_ensemble.append(reloaded_model1)
models_to_ensemble.append(reloaded_model2)
models_to_ensemble.append(reloaded_model3)






print("\n3. Creation of Ensemble Network (Soft Voting)...")


input_shape = X_test.shape[1:]
ensemble_input = tf.keras.Input(shape=input_shape)


ensemble_outputs = [model(ensemble_input) for model in models_to_ensemble]


y_avg = layers.Average()(ensemble_outputs)
ensemble_model = models.Model(inputs=ensemble_input, outputs=y_avg, name='Soft_Voting_Ensemble')





ensemble_predictions_prob = ensemble_model.predict(X_test, verbose=1)


ensemble_classes = np.argmax(ensemble_predictions_prob, axis=1)

final_accuracy = accuracy_score(y_test_flat, ensemble_classes)





for i, model in enumerate(models_to_ensemble):
    loss, acc = model.evaluate(X_test, y_test_one_hot, verbose=0)
    print(f"Test Accuracy Model {i+1}: {(acc):.4f}")


print(f"Test Accuracy of the FINAL ENSEMBLE: {final_accuracy:.4f} ")
```
